Question 1 : What are Type I and Type II errors in hypothesis testing, and how do they
impact decision-making?


In hypothesis testing, decisions are made about a population based on sample data. However, there is always a possibility of making a wrong decision. These mistakes are called Type I and Type II errors.

1. Type I Error (False Positive)

Definition:
A Type I error occurs when we reject a true null hypothesis (H‚ÇÄ).

Essentially, we think there is an effect when there is none.

Probability of Type I error = Œ± (significance level)

Example:

A drug company claims a new drug works.

Null hypothesis (H‚ÇÄ): The drug has no effect.

If we reject H‚ÇÄ but the drug actually has no effect, we made a Type I error.

Impact:

Leads to false alarms.

Can result in wasted resources or incorrect actions.

Question 2:What is the P-value in hypothesis testing, and how should it be interpreted
in the context of the null hypothesis?


P-value in Hypothesis Testing

Definition:
The P-value is the probability of observing a test statistic as extreme as, or more extreme than, the one calculated from the sample data, assuming the null hypothesis (H‚ÇÄ) is true.

In simple terms:

The P-value tells us how likely our observed data would occur if H‚ÇÄ were true.
Suppose a company tests a new sales strategy:

H‚ÇÄ: New strategy has no effect on sales

H‚ÇÅ: New strategy increases sales

After analyzing the data, the P-value = 0.03 and Œ± = 0.05

Since 0.03 < 0.05, we reject H‚ÇÄ, concluding the new strategy likely increases sales.

If P-value = 0.08:

Since 0.08 > 0.05, we fail to reject H‚ÇÄ, meaning evidence is insufficient to say the strategy works.

. What is a Confidence Interval (CI)?

A confidence interval is a range of values, derived from sample data, that is likely to contain the true population parameter (like the mean or proportion) with a certain level of confidence.

It gives not just a point estimate (e.g., sample mean) but a range that accounts for sampling variability.

The confidence level is usually expressed as a percentage, like 90%, 95%, or 99%, which indicates how confident we are that the interval contains the true population parameter.

Example:
If a 95% confidence interval for the average height of students is [160 cm, 170 cm], we are 95% confident that the true average height lies between 160 cm and 170 cm.

Question 5: Describe the purpose and assumptions of an ANOVA test. How does it
extend hypothesis testing to more than two groups?

Purpose of ANOVA

ANOVA (Analysis of Variance) is a statistical test used to compare the means of three or more groups to see if at least one group mean is significantly different from the others.

Instead of doing multiple t-tests (which increases the risk of Type I error), ANOVA tests all groups simultaneously.

It is widely used in experiments, surveys, or studies where there are more than two categories of a factor (independent variable).

Example:
You want to test whether the average test scores differ among students from three different teaching methods. ANOVA can determine if there is a statistically significant difference in mean scores among the three groups.

2. Assumptions of ANOVA

ANOVA relies on several key assumptions to be valid:

Independence of observations:

Data in each group should be collected independently of other groups.

Normality:

The dependent variable should be approximately normally distributed within each group.

Homogeneity of variance (Homoscedasticity):

The variance among the groups should be roughly equal.

Continuous dependent variable:

The outcome being tested should be on an interval or ratio scale (numeric).

If these assumptions are violated, results of ANOVA may not be reliable.

Question 6: Write a Python program to perform a one-sample Z-test and interpret the
result for a given dataset.


```python
import numpy as np
from scipy import stats

# Sample data: weights of 30 students (in kg)
sample_data = np.array([68, 72, 71, 69, 70, 67, 73, 70, 69, 71,
                        68, 70, 72, 69, 70, 71, 68, 69, 72, 70,
                        69, 71, 70, 68, 72, 69, 70, 71, 70, 69])

# Population parameters
population_mean = 70      # H0: mu = 70
population_std = 5        # Known population standard deviation
n = len(sample_data)      # Sample size

# Calculate sample mean
sample_mean = np.mean(sample_data)

# Compute Z-statistic
z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(n))

# Compute p-value (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

# Output results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-Statistic: {z_stat:.3f}")
print(f"P-Value: {p_value:.4f}")

# Interpretation
alpha = 0.05  # 5% significance level
if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from 70 kg.")
else:
    print("Fail to reject the null hypothesis: No significant difference from 70 kg.")
```

Question 7:Simulate a dataset from a binomial distribution (n = 10, p = 0.5) using
NumPy and plot the histogram.


```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the binomial distribution
n = 10      # Number of trials
p = 0.5     # Probability of success
size = 1000 # Number of samples to generate

# Simulate the dataset
binomial_data = np.random.binomial(n, p, size)

# Display first 10 values
print("First 10 simulated values:", binomial_data[:10])

# Plot the histogram
plt.figure(figsize=(8,5))
plt.hist(binomial_data, bins=range(n+2), edgecolor='black', alpha=0.7)
plt.title('Histogram of Binomial Distribution (n=10, p=0.5)')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.xticks(range(n+1))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Question 8: Generate multiple samples from a non-normal distribution and implement
the Central Limit Theorem using Python.


```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate a non-normal population (e.g., exponential distribution)
population_size = 100000
population_data = np.random.exponential(scale=2, size=population_size)

# Plot the population distribution
plt.figure(figsize=(8,4))
plt.hist(population_data, bins=50, color='skyblue', edgecolor='black')
plt.title("Population Distribution (Exponential - Non-Normal)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

# Step 2: Implement the Central Limit Theorem
sample_size = 50   # Number of observations per sample
num_samples = 1000 # Number of samples to draw
sample_means = []

# Generate multiple samples and compute their means
for _ in range(num_samples):
    sample = np.random.choice(population_data, sample_size)
    sample_means.append(np.mean(sample))

sample_means = np.array(sample_means)

# Step 3: Plot histogram of sample means
plt.figure(figsize=(8,4))
plt.hist(sample_means, bins=30, color='lightgreen', edgecolor='black')
plt.title("Sampling Distribution of Sample Means (CLT)")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Step 4: Print summary statistics
print(f"Population Mean: {np.mean(population_data):.2f}")
print(f"Sample Means Mean: {np.mean(sample_means):.2f}")
print(f"Sample Means Std Dev: {np.std(sample_means):.2f}")
```

Question 9: Write a Python function to calculate and visualize the confidence interval
for a sample mean.


```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def confidence_interval(data, confidence=0.95):
    """
    Calculate and visualize the confidence interval for a sample mean.

    Parameters:
        data (array-like): Sample data
        confidence (float): Confidence level (default 0.95)

    Returns:
        mean (float), lower_bound (float), upper_bound (float)
    """
    # Convert to numpy array
    data = np.array(data)
    n = len(data)
    mean = np.mean(data)
    std_err = np.std(data, ddof=1) / np.sqrt(n)  # sample standard error

    # t critical value
    t_crit = stats.t.ppf((1 + confidence) / 2, df=n-1)

    # Confidence interval
    margin_error = t_crit * std_err
    lower_bound = mean - margin_error
    upper_bound = mean + margin_error

    # Print results
    print(f"Sample Mean: {mean:.2f}")
    print(f"{int(confidence*100)}% Confidence Interval: [{lower_bound:.2f}, {upper_bound:.2f}]")

    # Visualization
    plt.figure(figsize=(6,4))
    plt.errorbar(x=0, y=mean, yerr=margin_error, fmt='o', color='blue', capsize=8, markersize=8)
    plt.xlim(-1, 1)
    plt.xticks([])
    plt.ylabel('Value')
    plt.title(f'{int(confidence*100)}% Confidence Interval for Sample Mean')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

    return mean, lower_bound, upper_bound

# -------------------------
# Example Usage
sample_data = [68, 72, 71, 69, 70, 67, 73, 70, 69, 71]
confidence_interval(sample_data, confidence=0.95)
```

Question 10: Perform a Chi-square goodness-of-fit test using Python to compare
observed and expected distributions, and explain the outcome.


```python
import numpy as np
from scipy.stats import chisquare

# Observed frequencies
observed = np.array([8, 10, 9, 11, 12, 10])

# Expected frequencies (fair die: equal probability)
expected = np.array([10, 10, 10, 10, 10, 10])  # total 60/6 = 10

# Perform Chi-square goodness-of-fit test
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

# Output results
print(f"Chi-square Statistic: {chi2_stat:.3f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The observed distribution differs significantly from the expected distribution.")
else:
    print("Fail to reject the null hypothesis: The observed distribution is consistent with the expected distribution.")
```
